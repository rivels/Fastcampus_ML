```python
from sklearn.model_selection import train_test_split

features = iris.data[:,2:]
labels = iris.target

X_train, X_test, y_train, y_test = train_test_split(features,
                                                   labels,
                                                   test_size=0.2,
                                                    stratify=labels,
                                                   random_state=13)
# stratify=labels이라는 옵션을 추가로 적용해서 더 정확하게 
# 테스트와 트레인을 나눌 수 있다
```

```python
np.unique(y_test, return_counts=True)
```

```python
from sklearn.tree import DecisionTreeClassifier

iris_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
iris_tree.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred_tr = iris_tree.predict(X_train)
accuracy_score(y_train, y_pred_tr)
```

```python
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions

plt.figure(figsize=(12,8))
plot_decision_regions(X=X_train, y=y_train, clf=iris_tree,
                     legend=2)
plt.show()
```

```python
y_pred_test = iris_tree.predict(X_test)
accuracy_score(y_test, y_pred_test)
```

```python
scatter_highlight_kwargs = {'s' : 150, 'label':'Test data',
                           'alpha':0.9}
scatter_kwargs = {'s':120, 'edgecolor' : None, 'alpha' : 0.7}

plt.figure(figsize=(12,8))
plot_decision_regions(X=features, y=labels,
                     X_highlight=X_test,clf=iris_tree,
                     legend=2,
                     scatter_highlight_kwargs=scatter_highlight_kwargs,
                     scatter_kwargs=scatter_kwargs,
                     contourf_kwargs={'alpha':0.2})
plt.show()
```

```python
import pandas as pd

data = {'x':[1.,2.,3.,4.,5.], 'y':[1.,3.,4.,6.,5.]}
df = pd.DataFrame(data)
df
```

```python
import statsmodels.formula.api as smf

lm_model = smf.ols(formula="y~ x ", data=df).fit()
```

```python
lm_model.params
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.lmplot(x='x',y='y',data=df);
```

```python
resid = lm_model.resid
resid
```

```python
import numpy as np
mu = np.mean(df.y)
y = df.y
yhat = lm_model.predict()
np.sum((yhat - mu)**2 / np.sum((y-mu)**2))
```

```python
lm_model.rsquared
```

```python
sns.distplot(resid,color='black');
```

```python
data = pd.read_csv('ecommerce.csv')
```

```python
data.head()
```

```python
# 필요없는 칼럼 제거
data.drop(['Email','Address','Avatar'], axis = 1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split

X = data.drop('Yearly Amount Spent', axis=1)
# 위의 코드는 'yearly ..'컬럼을 드롭하고 남은 컬럼들을 리턴함
y = data['Yearly Amount Spent']

X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                   test_size=0.3,
                                                   random_state=13)
```

```python
import statsmodels.api as sm

lm = sm.OLS(y_train, X_train).fit()
lm.summary()
```

```python
predictions = lm.predict(X_test)
predictions
```

```python
y_test
```

```python
sns.scatterplot(y_test, predictions);
plt.plot([300,800],[300,800],'r',ls='dashed',lw=3)
```

```python
from sklearn import metrics
print('MSE:',metrics.mean_squared_error(y_test,predictions))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
